```python
#Paquetes python para procesar datos json y csv
import json
import csv
```

```python
#Lista para almacenar las filas del csv transformada en json
lista_json = []
```

```python
#Funcion open, sentecia with cierra el archivo abierto automáticamente
#'r' modo de acceso al archivo, abrir un archivo solo para lectura
#utf-8 estandar para la correcta codificacion de caracteres para comunicacion electronica

with open('imdb_movies.csv','r',encoding='utf-8') as archivo_csv:
    
    #Modulo csv.DictReader crea un diccionario (claves y valores) por cada fila del archivo csv
    lector_csv = csv.DictReader(archivo_csv)
    
    #row: fila del archivo
    for row in lector_csv:
        lista_json.append(row)
```

```python
lista_json
```

```python
#Libreria python para el manejo de couchdb (base de datos)
import couchdb
```

```python
#Conectarse al servidor de couchdb
couch = couchdb.Server('http://admin:admin@localhost:5984/')
```

```python
#Crear una base de datos en couchdb
creardb = couch.create('peliculas')
```

```python
#BD creada almacenada en una variable
db = couch['peliculas']
```

```python
#Bucle que recorre la lista que contiene los elementos json y mediante el metodo save los va almacenando en al bd 'peliculas'
for doc in lista_json:
    db.save(doc)
```
